```python
import pandas as pd
import numpy as npd
```

```python
df_udemy = pd.read_csv('web_scrapping/Datasets/udemy_courses.csv')
```

```python
df_udemy.head()
```

```python
df_udemy.drop(['course_id','price','num_subscribers','num_reviews','num_lectures'],axis=1,inplace =True)
```

```python
df_udemy.head()
```

```python
df_udemy.rename(columns = {'is_paid':'certification', 'subject':'organization'}, inplace = True) 
df_udemy.drop(['published_timestamp'],axis=1,inplace =True)
```

```python
df_udemy.drop(['content_duration'],axis=1,inplace =True)
df_udemy.head()
```

```python
df_udemy['platform']='udemy'
```

```python
booleandf = df_udemy.select_dtypes(include=[bool])
booleanDictionary = {True: 'COURSE', False: 'FREE'}

for column in booleandf:
    df_udemy['certification'] = df_udemy['certification'].map(booleanDictionary)
```

```python
df_udemy.head()
```
